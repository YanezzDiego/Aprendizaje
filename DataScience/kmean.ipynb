# Generamos una base de datos artificial


```python
import matplotlib.pyplot as plt
import num

plt.jet()
import sklearn.datasets as datasets
X, Y = datasets.make_blobs(centers=4, cluster_std=0.5, random_state=0)

#la base de datos es una base con 2 variables independientes X y 1 variable dependiente Y
```

# Podemos visualizar como se distribuyen nuestros datos

```python
plt.scatter(X[:,0], X[:,1]); #0 se refiere a la primera columna y 1 a la segunda
```

```python
plt.scatter(X[:,0], X[:,1], c=Y); #esto es lo que el algoritmo k-means debera hacer
```

# Clusters k-mean

```python
from sklearn.cluster import KMeans
n=4
kmeans = KMeans(n, random_state=8)
Y_hat = kmeans.fit(X).labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=Y_hat); #se puede observar que los clusters se ajustan de buena manera a la realidad de los datos
```
