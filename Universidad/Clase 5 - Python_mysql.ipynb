```python
cnx = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "tQYCnFua9HvCN*E3",
    database = "football_db"
)

cursor = cnx.cursor()

query = 'SELECT * FROM football_db.appearances LIMIT 500;'

cursor.execute(query)

results = cursor.fetchall()

col_names = ['idappearances', 'goals', 'ownGoals', 'shots', 'xGoals', 
             'xGoalsChain', 'xGoalsBuildup', 'assist', 'keyPasses', 
             'xAssists', 'position', 'positionOrder', 'yellowCard', 
             'redCard', 'time', 'substituteIn', 'substituteOut', 
             'games_gamesID', 'players_playerID', 'leagues_leagueID']

data = pd.DataFrame(results, columns = col_names)

df = data.copy()

new_table = preprocessing_data(df)

for index,row in new_table.iterrows():
    tupla = tuple(row)
    query = f"INSERT INTO football_db.resumen(position,idappearances,goals,ownGoals,yellowCard,redCard,time) VALUE ('{tupla[0]}',{tupla[1]},{tupla[2]},{tupla[3]},{tupla[4]},{tupla[5]},{tupla[6]});"
    print("Insertando tupla: ",tupla)
    cursor.execute(query)

cnx.commit()
    
cnx.close()
```

```python
def preprocessing_data(df):
    df.loc[df['substituteIn']>1,'substituteIn'] = np.nan
    df.loc[df['substituteOut']>1,'substituteOut'] = np.nan

    df = df.dropna()

    new_table = df.groupby(by = ['position']).count()

    new_table = new_table.reset_index()

    new_table = new_table[['position','idappearances','goals','ownGoals','yellowCard','redCard','time']]
    
    return new_table
```
