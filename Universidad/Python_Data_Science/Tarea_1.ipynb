```python
#FUNCION DC
def DC(x,y):
  minimo = min(x,y)
  maximo = max(x,y)
  acumulador=0
  i=minimo

  while i<maximo:
    if i%2==0:
      acumulador= acumulador+i
      i=i+1
    else:
      i=i+1
  return acumulador

#FUNCION DF
def DF(x,y):
  minimo = min(x,y)
  maximo = max(x,y)

  numero=minimo
  resta=0
  aux=0

  while numero < maximo:
    while aux==0:
      if numero%2==0:
        resta=numero
        aux=aux+1
        numero=numero+1
      else:
        numero=numero+1
    if numero%2==0:
      resta= resta-numero
      numero=numero+1
    else:
      numero=numero+1
  return resta

x=int(input("Ingrese un valor para x "))
y=int(input("Ingrese un valor para y "))

print("MENU")
a=int(input("1) Ingrese 1 si desea calcular DC \n2) Ingrese 2 si desea calcular DF \n3) Ingrese 3 si desea salir del programa \n "))

while a<1 or a>3:
  a=int(input("Ingrese un menú válido.\n "))


if a==1:
  resultado= DC(x,y)
elif a==2:
  resultado= DF(x,y)
else:
  print("Ha salido del programa.")

print(resultado)
    
```
