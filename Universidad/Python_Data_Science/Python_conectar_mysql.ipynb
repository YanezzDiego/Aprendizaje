## Set Data Types

```python
df.isna().sum()
```

```python
# Date 
df["date"]= pd.to_datetime(df["date"])
```

```python
#variables numericas
no_numeric= ['date','street', 'city','statezip', 'country'] 
df_numeric= df.loc[:,df.columns.difference(no_numeric)]

for col_name,col in df_numeric.items():
    df[col_name]=pd.to_numeric(col,errors="coerce")
    
```

```python
#variables no numericas
df["street"]=df["street"].astype(str)
df["city"]=df["city"].astype(str)
df["statezip"]=df["statezip"].astype(str)
df["country"]=df["country"].astype(str)
```

```python
#Crear nuevas columnas
df["bathroomsXbedrooms"] = df["bedrooms"]*df["bathrooms"]
```
