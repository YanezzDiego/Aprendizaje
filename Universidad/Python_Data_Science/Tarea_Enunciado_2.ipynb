## Tarea Enunciado 2 : Synchronous Machine Data Set Data Set
Los motores sincrónicos (SMs) son motores de corriente alterna con velocidad constante. La base de datos es obtenida a partir del entorno de operaciones de un experimento real. La tarea es crear un modelo para estimar la corriente de excitación de los motores sincrónicos.



## Descripción de la base de datos
Los atributos presentes en la base de datos son los siguientes:


1.    **ly:** Corriente de carga (variable independiente)
2.    **PF:** Factor de potencia (variable independiente)
3.    **e:** Error del factor de potencia (variable independiente)
4.    **dIf** Cambio de la corriente de excitación de la máquina síncronica (variable independiente)
5.    **If**: Corriente de excitación de la máquina síncrona (variable dependiente)

```python
# Load CSV using Pandas
import pandas as pd
filename = "/content/gdrive/MyDrive/ClaseProcesamiento/data/synchronous_machine.csv"
df = pd.read_csv(filename,sep=";")
print(df)
```

## Librerias

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

## Dividir X_reg y Y_reg en conjuntos de test y train 
Primero, separamos la base de datos en 70% train y 30% test

```python
train, test = train_test_split(df, test_size = 0.30)
print("Ejemplos usados para train: ", len(train))
print("Ejemplos usados para test: ", len(test))
```

Luego, obtenemos X_reg_train, Y_reg_train, X_reg_test y Y_reg_test

```python
#X_reg_traing e Y_reg_train
array_train= train.values
x_reg_train= array_train[:,0:4]
y_reg_train= array_train[:,4]
print("Los valores X usados para train son: \n ", x_reg_train,"\n")
print("----------------------------------------------------------------------")
print("Los valores Y usados para train son: \n ", y_reg_train, "\n")
```

```python
#X_reg_test e Y_reg_test
array_test= test.values
x_reg_test= array_test[:,0:4]
y_reg_test= array_test[:,4]
print("Los valores X usados para test son: \n ", x_reg_test,"\n")
print("------------------------------------------------------------------------")
print("Los valores Y usados para test son: \n ", y_reg_test, "\n")
```

## Regresiones 
Se entrenan los modelos con el 70% y se prueban sobre el otro 30% de la base de datos

### Regresión Lineal


```python
# Linear Regression
from sklearn.linear_model import LinearRegression

kfold = KFold(n_splits = 10)
model = LinearRegression()
scoring = "neg_mean_squared_error"
results = cross_val_score(model, x_reg_train, y_reg_train, cv = kfold, scoring = scoring)
print("El MSE para LiR es "+str(results.mean()))
print("-----------------------------------------------------")
model.fit(x_reg_train, y_reg_train)
y_predict = model.predict(x_reg_test)
from sklearn.metrics import r2_score

print('R2_Score: ', r2_score(y_reg_test, y_predict))
print("El modelo con todas las variables introducidas es capaz de explicar el ",r2_score(y_reg_test, y_predict)*100,"% de la variabilidad observada en la variable dependiente CORRIENTE DE EXCITACION DE LA MAQUINA SINCRONICA  \n ")
print("El intercepto del modelo es ", model.intercept_)
print("Los coeficientes del modelo son ", model.coef_ ," respectivamente" )
```

### Ridge Regression

```python
# Ridge 
from sklearn.linear_model import Ridge

kfold = KFold(n_splits = 10)
model = Ridge()
scoring = "neg_mean_squared_error"
results = cross_val_score(model, x_reg_train, y_reg_train, cv = kfold, scoring = scoring)
print("El MSE para LiR es "+str(results.mean()))
print("-----------------------------------------------------")
model.fit(x_reg_train, y_reg_train)
y_predict = model.predict(x_reg_test)
from sklearn.metrics import r2_score

print('R2_Score: ', r2_score(y_reg_test, y_predict))
print("El modelo con todas las variables introducidas es capaz de explicar el ",r2_score(y_reg_test, y_predict)*100,"% de la variabilidad observada en la variable dependiente CORRIENTE DE EXCITACION DE LA MAQUINA SINCRONICA  \n ")
print("El intercepto del modelo es ", model.intercept_)
print("Los coeficientes del modelo son ", model.coef_ ," respectivamente" )
```
